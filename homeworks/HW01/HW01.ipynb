```python
import numpy as np
```

```python
arr_list = np.array([1, 2, 3, 4]) # из списка в массив NumPy
arr_list
```

```python
arr_zeros = np.zeros(10) # массив из 10 нулевых элементов
arr_zeros
```

```python
arr_arange = np.arange(1, 10, 0.1) # генерация массива последовательности в диапазоне [1; 10) с шагом 0.1
arr_arange
```

```python
arr_arange.shape # размерность массива, его "форма"
```

```python
arr_arange.dtype # тип данных элементов массива - float64
```

```python
matrix = np.arange(1, 10, 0.1).reshape(9, 10) # изменение "формы" массива и превращение в матрицу
matrix
```

```python
matrix[1:3] # первая и вторая строки матрицы
```

```python
matrix[1, 3:5] # третий и четвертый элементы первой строки
```

```python
matrix[1, 1] # первый элемент первой строки
```

```python
a = np.array([1, 2, 3])
b = np.array([1, 2, 3]).reshape(3, 1)
a @ b # произведение векторов (по сути, скалярное произведение)
```

```python
a + b.T # сумма двух векторов
```

```python
a.mean() # среднее арифметическое
```

```python
np.median(a) # медиана
```
